````python

import os
import json
import numpy as np
import librosa
import tensorflow as tf
import tensorflow_hub as hub
import whisper
import torch
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from transformers import pipeline



device = 0 if torch.cuda.is_available() else -1



def translate_to_english(text):
    prompt = f"""
Translate the following text into English.
Return only translated sentence.

Text:
{text}
"""
    response = llm.invoke(prompt)
    return response.content.strip()



class AudioPreprocessor:
    def __init__(self, target_sr=16000):
        self.target_sr = target_sr

    def load(self, path):
        audio, sr = librosa.load(path, sr=self.target_sr, mono=True)

        if np.max(np.abs(audio)) > 0:
            audio = audio / np.max(np.abs(audio))

        duration = len(audio) / sr
        return audio, sr, duration

    def is_silence(self, audio, threshold=0.001):
        return np.mean(np.abs(audio)) < threshold


class AudioClassifier:
    def __init__(self):
        self.model = hub.load("https://tfhub.dev/google/yamnet/1")

    def classify(self, audio):
        scores, _, _ = self.model(tf.cast(audio, tf.float32))
        mean_scores = np.mean(scores.numpy(), axis=0)

        speech_score = float(np.max(mean_scores[:20]))
        music_score = float(np.max(mean_scores[130:150]))

        if speech_score > 0.2 and music_score > 0.2:
            mode = "both"
        elif speech_score > 0.2:
            mode = "speech_only"
        elif music_score > 0.2:
            mode = "music_only"
        else:
            mode = "neither"

        return mode

class SpeechBranch:
    def __init__(self):
        self.whisper = whisper.load_model("medium")

        self.emotion_model = pipeline(
            "audio-classification",
            model="ehcalabres/wav2vec2-lg-xlsr-en-speech-emotion-recognition",
            device=device
        )

    def process(self, audio_path):
        result = self.whisper.transcribe(audio_path)
        transcript = result["text"]
        language = result.get("language", "unknown")

        if language != "en":
            translated = translate_to_english(transcript)
        else:
            translated = transcript

        emotion_result = self.emotion_model(audio_path)
        dominant = max(emotion_result, key=lambda x: x["score"])

        return {
            "original_transcript": transcript,
            "translated_transcript": translated,
            "language": language,
            "speech_emotion": dominant["label"],
            "emotion_confidence": float(dominant["score"])
        }



class MusicBranch:
    def __init__(self):
        self.model = pipeline(
            "audio-classification",
            model="superb/wav2vec2-base-superb-er",
            device=device
        )

    def process(self, audio_path):
        results = self.model(audio_path)
        dominant = max(results, key=lambda x: x["score"])

        return {
            "music_emotion": dominant["label"],
            "emotion_confidence": float(dominant["score"])
        }


class LLMAnalyzer:
    def analyze(self, structured_features):

        prompt = f"""
You are a strict JSON-only classifier for psychological manipulation detection.

TASK:
Analyze the structured audio input and determine if it contains psychological manipulation.

INPUT:
audio_mode: {structured_features["audio_mode"]}
duration_seconds: {structured_features["duration_seconds"]}
music_features: {structured_features["music_features"]}
speech_features: {structured_features["speech_features"]}

DECISION RULES:
1. Classify as "manipulative": true ONLY if there is clear, strong, and unambiguous evidence.
2. If signals are weak, indirect, or conflicting → return false.
3. Do NOT infer beyond given features.
4. Be conservative in classification.

OUTPUT REQUIREMENTS (CRITICAL):
- Return ONLY valid JSON.
- Do NOT include explanations outside JSON.
- Do NOT include markdown, backticks, or the word "json".
- Do NOT wrap output in any key like "result".
- Output must be directly parseable using json.loads().

STRICT JSON SCHEMA:
{{
  "manipulative": boolean,
  "confidence": float (0 to 1),
  "technique": "guilt_tripping" | "fear_mongering" | "gaslighting" | "emotional_coercion" | "deception" | "none",
  "reasoning": string,
  "evidence_strength": "strong" | "moderate" | "weak"
}}

FINAL RULE:
If uncertain → return:
{{
  "manipulative": false,
  "confidence": 0.5,
  "technique": "none",
  "explanation": "Insufficient evidence of manipulation",
  "evidence_strength": "weak"
}}

RETURN JSON ONLY.
"""
        
    

        try:
            response=llm.invoke(prompt).content
            return {"result": response}
        except:
            return {
                "manipulative": False,
                "manipulation_type": "unknown",
                "confidence": 0.5,
                "explanation": raw,
                "response_strategy": "Could not parse model output."
            }


class AudioPipeline:
    def __init__(self):
        self.pre = AudioPreprocessor()
        self.router = AudioClassifier()
        self.speech = SpeechBranch()
        self.music = MusicBranch()
        self.llm = LLMAnalyzer()

    def analyze(self, audio_path):
        print("audio-start")
        audio, sr, duration = self.pre.load(audio_path)

        if self.pre.is_silence(audio):
            return {"audio_mode": "silence"}

        mode = self.router.classify(audio)

        speech_data = None
        music_data = None

        if mode in ["speech_only", "both"]:
            speech_data = self.speech.process(audio_path)

        if mode in ["music_only", "both"]:
            music_data = self.music.process(audio_path)

        structured_features = {
            "audio_mode": mode,
            "duration_seconds": round(duration, 2),
            "speech_features": speech_data,
            "music_features": music_data
        }
        print("audio-reasoning")
        llm_verdict = self.llm.analyze(structured_features)
        print("audio-reasoned")
        return {
            "features": structured_features,
            "llm_verdict": llm_verdict
        }

    def analyze_2(self, audio_path):

        audio, sr, duration = self.pre.load(audio_path)
    
        if self.pre.is_silence(audio):
            return {"audio_mode": "silence"}
    
        mode = self.router.classify(audio)
    
        speech_data = None
        music_data = None
    
        if mode in ["speech_only", "both"]:
            speech_data = self.speech.process(audio_path)
    
        if mode in ["music_only", "both"]:
            music_data = self.music.process(audio_path)
    
        structured_features = {
                "audio_mode": mode,
                "duration_seconds": round(duration, 2),
                "speech_features": speech_data,
                "music_features": music_data
            }
        return structured_features 

````
